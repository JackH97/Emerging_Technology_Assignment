```python
#required imports for the following task

#Iris data set upload
from sklearn import datasets

# Numerical arrays.
import numpy as np

# Machine learning - KMeans.
from sklearn.cluster import KMeans

# Plotting.
import matplotlib.pyplot as plt

#used for reading in files
import pandas as pd

# load the Iris flowers dataset
iris_data = datasets.load_iris()
```

```python
#Takes dataset and outputs the info for reading
Table = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)

Table
```

```python
# Target
X = iris_data.data[:, :2]
y = iris_data.target
```

```python
# data outputted using a scatterplot.
plt.scatter(X[:,0], X[:,1], c=y, cmap='jet')
plt.xlabel('SPEAL LENGTH', fontsize=10)
plt.ylabel('SEPAL WIDTH', fontsize=10)
```

```python

# apply k-means clustering to Fisher’s famous Iris data set.
kmns = KMeans(n_clusters = 4, n_jobs = 6, random_state=25)
kmns.fit(X)
```

```python
# finding and outputing the center points.
mid = kmns.cluster_centers_
print(mid)
```

```python
# comparing the first scatterplot against the clustered findings
fig, axes = plt.subplots(1, 2, figsize=(16,8))

axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap="rainbow")
axes[0].set_xlabel("Sepal Length", fontsize=15)
axes[0].set_ylabel("Sepal Width", fontsize=15)
axes[0].tick_params(direction="in", length=8, width=4, colors='k', labelsize=15)
axes[0].set_title("Original Scatterplot", fontsize=20)

axes[1].scatter(X[:, 0], X[:, 1], c=kmns.labels_,  cmap="gist_rainbow")
axes[1].set_xlabel("Sepal Length", fontsize=15)
axes[1].set_ylabel("Sepal Width", fontsize=15)
axes[1].tick_params(direction='in', length=8, width=4, colors='k', labelsize=15)
axes[1].set_title("Clustered Scatterplot", fontsize=20)
```
