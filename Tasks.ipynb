```python
def sqrt2():
    # X is declared a variable
    x = 2
    # Squares the "x" variable with the equation
    y = x ** .5
    # Outputs the square root of 2 with the variable "y"
    print('The square root of the number 2 is %0.2f'%(y))
    
    # I use variable called 'g' for the math term googol for 10 to the power of 100
    # and with doing this i multiply then to get the value to be pushed to the left
    # side of the decimal place. Then i have a variable called 'store' to hold the number 0
    # to have better accuracy in the while loop
    g = 10**100
    sd = x *(g**2)
    store = 0
    
    # A while loop to increase the accuracy of the number being squared 
    while store != g:
        # Setting the variable 'store' to be the same as the variable "g"
        store = g
        # Here is where we do division for a better result
        g = (store + (sd//store)) >> 1
        
    # Here i convert my "g" variable with the value to a string 
    store2 = list(str(g))
    
    # searches the 'store2' variable that has stored the value of the variable 'g' above for 
    # the number after as its the first number after the decimal point in the square root of 2
    store2.insert(1, '.')
    
    # Then when the decimal point is inserted after the number 1
    # we then attach the rest of the variable 'store2' 
    # to the to give us the square root of 2 to 100 decimal places
    out_sqrt = ''.join(store2)
    
    # Prints out the square root of 2 to 100 decimal places
    print('the square root of 2 to 100 decimal places is:', out_sqrt)
```

### The Differences between both:

STDEV.P function is used when your data represents the entire population. For the squared deviation 
it is divided by the total number of sample -1. It is represented as N.
<img src="Images/STDEVp.png">

STDEV.S function is used when your data is a sample of the entire population. For the squared deviation,
it is divided by the total number of arguments, mostly represented as N-1.
<img src="Images/STDEVs.png">

To explain differences between both: https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/variance-standard-deviation-sample/a/population-and-sample-standard-deviation-review#:~:text=Standard%20deviation%20measures%20the%20spread%20of%20a%20data%20distribution.&text=If%20the%20data%20is%20being,n%2D1%20n%E2%88%921%20.

2. Figuring out the functions work in microsoft excel: https://www.exceltip.com/statistical-formulas/how-to-use-stdev-s-function-in-excel.html  +  https://www.exceltip.com/statistical-formulas/how-to-use-excel-stdev-p-function.html

3. Understanding numpy and using it for the task: https://numpy.org/doc/stable/reference/generated/numpy.std.html

```python
import numpy as np

population = np.array([3,7,9,12,15,18,21,27,29,33,35,36,42,45,47,51,54,58])
sample = np.array([9,18,27,36,45,54])
 
#STDEV.P with population data
pop = np.sqrt(np.sum((population - np.mean(population))**2)/len(population))
# STDEV.P with sample data
STDEVp = np.sqrt(np.sum((sample - np.mean(sample))**2)/len(sample)) 
# STDEV.S with sample data
STDEVs = np.sqrt(np.sum((sample - np.mean(sample))**2)/len(sample)-1) 

print("Standard Deviation - Population: ", pop)
print("Standard Deviation.P - sample: ", STDEVp)
print("Standard Deviation.S - sample: ", STDEVs)
```

```python
#required imports for the following task

#Iris data set upload
from sklearn import datasets

# Numerical arrays.
import numpy as np

# Machine learning - KMeans.
from sklearn.cluster import KMeans

# Plotting.
import matplotlib.pyplot as plt

#used for reading in files
import pandas as pd

# load the Iris flowers dataset
iris_data = datasets.load_iris()
```

```python
#Takes dataset and outputs the info for reading
Table = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)

Table
```

```python
# Target
X = iris_data.data[:, :2]
y = iris_data.target
```

```python
# data outputted using a scatterplot.
plt.scatter(X[:,0], X[:,1], c=y, cmap='jet')
plt.xlabel('Speal Length', fontsize=20)
plt.ylabel('Sepal Width', fontsize=20)
```

```python
#apply k-means clustering to Fisherâ€™s famous Iris data set.
cluster = KMeans(n_clusters = 5, n_jobs = 5, random_state=24)
cluster.fit(X)
```

```python
# finding and outputing the center points.
mid = cluster.cluster_centers_
print(mid)
```

```python
# comparing the first scatterplot against the clustered findings
fig, axes = plt.subplots(1, 2, figsize=(16,8))

#Scatterplot 1 - Original
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap="rainbow", s=100)
axes[0].set_xlabel("Sepal Length", fontsize=15)
axes[0].set_ylabel("Sepal Width", fontsize=15)
axes[0].tick_params(direction="in", length=12, width=6, colors='k', labelsize=15)
axes[0].set_title("Original Scatterplot", fontsize=20)

#Scatterplot 2 - Clustered
axes[1].scatter(X[:, 0], X[:, 1], c=cluster.labels_,  cmap="gist_rainbow", s=100)
axes[1].set_xlabel("Sepal Length", fontsize=20)
axes[1].set_ylabel("Sepal Width", fontsize=20)
axes[1].tick_params(direction='in', length=12, width=6, colors='k', labelsize=15)
axes[1].set_title("Clustered Scatterplot", fontsize=20)
```
